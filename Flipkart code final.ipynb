# Importing libraries and ignoring limitations of rows and columns in data.

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
from bs4 import BeautifulSoup
import requests
import matplotlib.pyplot as mpy
import numpy as np
import seaborn as sb
import os

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth',None)
```

# Fetching data from flipkart.

```python
mname=[]
mprice=[]
mrating=[]
page_number=input("Enter the no of pages: ")
for i in range(1,int(page_number)+1):
    link="https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_1_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_1_0_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=mobiles&requestId=64f4337d-b7d2-4e8d-98c5-4772fcf4d070&p%5B%5D=facets.rating%255B%255D%3D1%25E2%2598%2585%2B%2526%2Babove&page="+str(i)
    #requesting server for accessing the link content
    request=requests.get(link)
    
    #for fetching the contents of the link
    content=BeautifulSoup(request.content, 'html.parser')
    
    #Finding the content under specific class
    mobname=content.find_all('div',{'class':'_4rR01T'})
    mobprice=content.find_all('div',{'class':'_30jeq3 _1_WHN1'})
    mobrating=content.find_all('div',{'class':'_3LWZlK'})
    
    print("Phone in page: "+str(i)) #print the no of pages
    
    print(len(mobname)) #print the total no of phones in each page
    
    #storing the fetched values in the empty lists.
    for i in (mobname):
        mname.append(i.text)
    for i in (mobprice):
        mprice.append(i.text)
    for i in (mobrating):
        mrating.append(i.text)
```

# Converting data to dataframe and saving the file in csv format.

```python
values={'mobile':mname,'price':mprice,'rating':mrating}
df=pd.DataFrame(values)
print(df)
df.to_csv("Flipkart_data.csv")
```

# Reading the csv file and viewing the file.

```python
df = pd.read_csv("Flipkart_data.csv")
df
```

# Checking NA values in the data.

```python
df.isna().sum()
```

# Using 'get' function to fetch the 1st word of the 'mobile'  column and storing it in 'mobile_company' column.

```python
df['mobile_company'] = df.mobile.str.split().str.get(0)
#Using the replace function to replace the names of some phone companies.
df.replace({'mobile_company' : { 'I' : "I Kall", 'APPLE' : "Apple" , "REDMI" : "Mi","Redmi" : "Mi", "LAVA" : "Lava" }},inplace = True)
df
```

# Using 'get' function to fetch the 2nd last word of the 'mobile' column and storing it in 'mobile_storage' column.

```python
df['mobile_storage'] = df.mobile.str.split().str.get(-2)
#Assigning NAN values to products not having storage info.
storage = ["16","32","64","128","256"]
df['mobile_storage'][~df['mobile_storage'].isin(storage)] = np.nan
df['mobile_storage'] = pd.to_numeric(df['mobile_storage'])
df
```

# Using 'lambda' function to fetch the values between "(" and  "," and storing in 'mobile_color' column.

```python
df['mobile_color'] = df['mobile'].apply(lambda x: x[x.find("(")+1:x.find(",")])
#Using 'get' function to fetch the last word between "(" and "," 
df['mobile_color'] = df['mobile_color'].str.split().str.get(-1)
#Getting the counts of different colors.
df['mobile_color'].value_counts()
```

# Assigning NAN values to products not having any color info.

```python
colors = ["Blue","Black","Purple","Green","Silver","White","Grey","Gold","Red","Yellow","Gray","Violet","Orange","Cream",
          "Greener","SeaBlue"]
#Assigning NAN values to products not having any color info.
df['mobile_color'][~df['mobile_color'].isin(colors)] = np.nan
#Replacing some color names to generic names.
df.replace({'mobile_color' : { 'SeaBlue' : "Blue", 'Greener' : "Green", 'Gray' : "Grey" }},inplace = True)
df.head(20)
```

# Removing '₹' from price column and setting it to integer type.

```python
df['price'] = df['price'].replace('[\₹\,]', '', regex=True).astype(int)
df.head()
```

# Deleting columns and setting the order of columns.

```python
df.drop(['Unnamed: 0', 'mobile'], axis = 1,inplace =True)
df = df[["mobile_company","price","mobile_color","mobile_storage","rating"]]
df.head()
```

### Histogram

```python
ratings = [2.5,3,3.5,4,4.5,5]
mpy.hist(df.rating, bins=ratings, color="#32a8a6",rwidth = 0.8)
mpy.xticks(ratings)
mpy.yticks([0,50,100,150,200,250,300,350])
mpy.ylabel("No. of phones")
mpy.xlabel("Phone ratings")
mpy.title("Distribution of No of phones w.r.t phone ratings")
mpy.show()
```

# Pie-chart

```python
Low=df.loc[df.price<10000].count()[0]
Low_medium=df.loc[(df.price>=10000) & (df.price<20000)].count()[0]
Medium=df.loc[(df.price>=20000) & (df.price<30000)].count()[0]
Medium_high=df.loc[(df.price>=30000) & (df.price<50000)].count()[0]
High=df.loc[(df.price>=50000) & (df.price<80000)].count()[0]
Premium=df.loc[df.price>80000].count()[0]
df.price=df['price']
phone_range=[Low,Low_medium,Medium,Medium_high,High,Premium]
labels=['Under 10000','Between 10000-19999','Between 20000-29999','Between 30000-49999','Between 50000-79999','Above 80000']
colors=['purple','yellow','red','green','orange','grey']
explode=(.1,.1,.1,.1,.6,.6)
mpy.title('Phones in different price ranges!')
mpy.pie(phone_range,colors=colors,autopct='%.1f %%', pctdistance=0.8, explode=explode,radius = 1.2,
       shadow = True)
mpy.legend(labels=labels,bbox_to_anchor = (1.1, 1.1))
mpy.show()
```

# Box and whiskers plot

```python
below_three=df.loc[df.rating<4]['price']
above_three=df.loc[df.rating>4]['price']
labels=['Below 4 rating','Above 4 rating']
#To increase or dicrease the boxplot size in the x or y direction.
mpy.figure(figsize=(12,14))
#changing outline and body color.'facecolor' is for body color.To enable facecolor we need to call 'patch_artist'.
#'medianprops' is required to match the line width inside the box with the border line width.
boxes=mpy.boxplot([below_three,above_three],labels=labels,medianprops={'linewidth':3},patch_artist=True)
for box in boxes['boxes']:
    box.set(color='#e303fc',facecolor='grey',linewidth=3)    
mpy.show()
```
